  2. **IRIS Dataset using sklearn** * Import IRIS data from sklearn or UCI. * Compute: * Mean, mode, median, standard deviation, confidence interval, standard error for each feature. * Compute correlation coefficients between each pair of features and plot heatmap. * Find covariance between length of sepal and petal. * Build contingency table for class feature.


```python
from sklearn.datasets import load_iris
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
print(df.head())
#compute mean ,median,mode,standard deviation, standard error
print("mean\n",df.mean())
print("mode\n",df.mode())
print("median\n",df.median())
print("std\n",df.std())
print("standard error \n",df.sem())

#compute correlation between each pair
cor = df.corr()
sns.heatmap(cor, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()


#covariance between sepal length (cm) and  petal length (cm)
print("covariance between sepal length (cm) and  petal length (cm) is",df["sepal length (cm)"].cov(df["petal length (cm)"]))
```
